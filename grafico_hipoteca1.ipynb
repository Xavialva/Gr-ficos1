```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df='/content/drive/MyDrive/PTDataScience/VisualDatos/mortgages (4).csv'
```

```python
df= pd.read_csv(df)
```

```python
df
```

```python
df.loc[(df['Mortgage Name']=='30 Year')&(df['Interest Rate']== 0.03)]
```

```python
df1=df.loc[(df['Mortgage Name']=='30 Year')&(df['Interest Rate']== 0.03)]
```

```python
df1['Interes acumulado']= df1['Interest Paid'].cumsum()
```

```python
df1
```

```python
df2=df.loc[(df['Mortgage Name']=='30 Year')&(df['Interest Rate']== 0.05)]
```

```python
df2['Interes acumulado']=df2['Interest Paid'].cumsum()
```

```python
df2
```

```python

plt.figure(figsize= (10, 5))
plt.plot(df1['Month'],df1['Interes acumulado'],c='b')
plt.plot(df2['Month'],df2['Interes acumulado'],c='k')
plt.xlabel('Meses')
plt.ylabel('Interes Acumulado')
plt.show()
```
